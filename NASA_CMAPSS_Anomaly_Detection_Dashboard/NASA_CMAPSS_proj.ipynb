```python
#!pip3 install scikit-learn

#Anomaly Detection with Isolation Forest
from sklearn.preprocessing import StandardScaler
```

```python
#remove sensor readings with less than 10 distinct values.
new_sensor_idx = [f"sensor{i}" for i in range(1,22)]
for i in range(1,22):
    if train_df1[f"sensor{i}"].nunique() < 10:
        train_df1.drop(axis = 1, columns = f"sensor{i}", inplace = True)
        new_sensor_idx.remove(f"sensor{i}")
```

```python
train_sensor_data = train_df1[new_sensor_idx]
display(train_sensor_data)

test_sensor_data = joined_test_df[new_sensor_idx]
display(test_sensor_data)
```

```python
my_scaler = StandardScaler()
train_data = my_scaler.fit_transform(train_sensor_data)
test_data = my_scaler.transform(test_sensor_data)
display(train_data)
display(test_data)

type(train_data)
type(test_data)
```

```python
from sklearn.ensemble import IsolationForest

isf_model = IsolationForest(n_estimators = 100, contamination = "auto", random_state = 1)
isf_model.fit(train_data)

train_ano_cnts = isf_model.predict(train_data)
test_ano_cnts = isf_model.predict(test_data)
```

```python
# !pip3 install seaborn

import seaborn as sns

plt.figure(figsize=(10,5))
sns.histplot(train_ano_cnts, bins=50, kde=True, label="Train")
sns.histplot(test_ano_cnts, bins=50, kde=True, color="red", label="Test")
plt.xlabel("Anomaly Score")
plt.ylabel("Density")
plt.title("Distribution of Anomaly Scores")
plt.legend()
plt.show()
```

```python
train_cnt,test_cnt = 0, 0
for tr in train_ano_cnts:
    if tr == -1:
        train_cnt += 1
for te in test_ano_cnts:
    if te == -1:
        test_cnt += 1
        
print(f"[Train Dataset]\nThe number of anomalies detected in the train dataset is {train_cnt} out of {len(train_ano_cnts)}.\nThe percentage of anomalies detected is {round(train_cnt*1.0/len(train_ano_cnts),4)}%\n\n")
print(f"[Test Dataset]\nThe number of anomalies detected in the test dataset is {test_cnt} out of {len(train_ano_cnts)}.\nThe percentage of anomalies detected is {round(test_cnt*1.0/len(test_ano_cnts),4)}%")
```

```python
sns.countplot(x=train_ano_cnts, color="skyblue")
plt.title("Anomaly Distribution in Train Data")
plt.show()

sns.countplot(x=test_ano_cnts, color = "blue")
plt.title("Anomaly Distribution in Test Data")
plt.show()
```

```python
sns.boxplot(x=train_ano_cnts, y=train_df1["RUL"])
plt.title("Anomaly vs. Remaining Useful Life (RUL)")
plt.show()
```
